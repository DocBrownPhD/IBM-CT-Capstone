Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
def marker_color(row):
    return 'green' if row['class'] == 1 else 'red'

spacex_df['marker_color'] = spacex_df.apply(marker_color, axis=1)
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Define coordinates for all launch sites and nearby features

# CCAFS SLC-40
lat_40, lon_40 = 28.56271, -80.58969
lat_40_ecoastline, lon_40_ecoastline = 28.56271, -80.56275
lat_40_wcoastline, lon_40_wcoastline = 28.56271, -80.5881
lat_40_railway, lon_40_railway = 28.56271, -80.58586
lat_40_highway, lon_40_highway = 28.56271, -80.5697
lat_40_city, lon_40_city = 28.3922, -80.6077  # Cape Canaveral

# CCAFS SLC 40
lat_S40, lon_S40 = 28.56226, -80.57535
lat_S40_ecoastline, lon_S40_ecoastline = 28.56226, -80.56859
lat_S40_wcoastline, lon_S40_wcoastline = 28.56226, -80.58954
lat_S40_railway, lon_S40_railway = 28.56226, -80.58625
lat_S40_highway, lon_S40_highway = 28.56226, -80.5689
lat_S40_city, lon_S40_city = 28.3922, -80.6077  # Cape Canaveral

# KSC LC 39A
lat_39A, lon_39A = 28.57327, -80.64494
lat_39A_ecoastline, lon_39A_ecoastline = 28.57327, -80.60692
lat_39A_wcoastline, lon_39A_wcoastline = 28.57327, -80.71731
lat_39A_railway, lon_39A_railway = 28.57327, -80.64324
lat_39A_highway, lon_39A_highway = 28.57327, -80.64326
lat_39A_city, lon_39A_city = 28.3772, -80.6691  # Merritt Island City

# VAFB SLC-4E
lat_4E, lon_4E = 34.63278, -120.61745
lat_4E_wcoastline, lon_4E_wcoastline = 34.63278, -120.62645
lat_4E_railway, lon_4E_railway = 34.63278, -120.62504
lat_4E_highway, lon_4E_highway = 34.63912, -120.47555
lat_4E_city, lon_4E_city = 34.6392, -120.4579  # Lompoc

# Calculate distances
distances = {
    "CCAFS SLC-40": {
        "East Coastline": calculate_distance(lat_40, lon_40, lat_40_ecoastline, lon_40_ecoastline),
        "West Coastline": calculate_distance(lat_40, lon_40, lat_40_wcoastline, lon_40_wcoastline),
        "Railway": calculate_distance(lat_40, lon_40, lat_40_railway, lon_40_railway),
        "Highway": calculate_distance(lat_40, lon_40, lat_40_highway, lon_40_highway),
        "City": calculate_distance(lat_40, lon_40, lat_40_city, lon_40_city),
    },
    "CCAFS SLC 40": {
        "East Coastline": calculate_distance(lat_S40, lon_S40, lat_S40_ecoastline, lon_S40_ecoastline),
        "West Coastline": calculate_distance(lat_S40, lon_S40, lat_S40_wcoastline, lon_S40_wcoastline),
        "Railway": calculate_distance(lat_S40, lon_S40, lat_S40_railway, lon_S40_railway),
        "Highway": calculate_distance(lat_S40, lon_S40, lat_S40_highway, lon_S40_highway),
        "City": calculate_distance(lat_S40, lon_S40, lat_S40_city, lon_S40_city),
    },
    "KSC LC 39A": {
        "East Coastline": calculate_distance(lat_39A, lon_39A, lat_39A_ecoastline, lon_39A_ecoastline),
        "West Coastline": calculate_distance(lat_39A, lon_39A, lat_39A_wcoastline, lon_39A_wcoastline),
        "Railway": calculate_distance(lat_39A, lon_39A, lat_39A_railway, lon_39A_railway),
        "Highway": calculate_distance(lat_39A, lon_39A, lat_39A_highway, lon_39A_highway),
        "City": calculate_distance(lat_39A, lon_39A, lat_39A_city, lon_39A_city),
    },
    "VAFB SLC-4E": {
        "West Coastline": calculate_distance(lat_4E, lon_4E, lat_4E_wcoastline, lon_4E_wcoastline),
        "Railway": calculate_distance(lat_4E, lon_4E, lat_4E_railway, lon_4E_railway),
        "Highway": calculate_distance(lat_4E, lon_4E, lat_4E_highway, lon_4E_highway),
        "City": calculate_distance(lat_4E, lon_4E, lat_4E_city, lon_4E_city),
    },
}
```

```python
# Create a base map centered at the average location of all sites
map_center = [36.0, -96.0]
site_map = folium.Map(location=map_center, zoom_start=5)

# Create a function to add markers and lines for each launch site
def add_markers_lines(site_map, lat, lon, features, site_name):
    marker = folium.Marker(
        [lat, lon],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % site_name,
            )
        )
    site_map.add_child(marker)
    
    for name, coords in features.items():
        lat_dest, lon_dest, distance = coords
        distance_marker = folium.Marker(
            [lat_dest, lon_dest],
            icon=DivIcon(
                icon_size=(20,20),
                icon_anchor=(0,0),
                html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
                )
            )
        site_map.add_child(distance_marker)

        line = folium.PolyLine(locations=[[lat, lon], [lat_dest, lon_dest]], weight=1)
        site_map.add_child(line)
    return site_map
```
